`Labels`

```python
xpoints = np.array([0,2,4,6,8]) 
ypoints = np.array([0,5,5,2,10]) 

plt.plot(xpoints, ypoints)

plt.title("Sports Watch Data")
plt.xlabel('No of Students')
plt.ylabel('Total Score')

plt.show()
```

`Font` Properties

```python
xpoints = np.array([0,2,4,6,8]) 
ypoints = np.array([0,5,5,2,10]) 

plt.plot(xpoints, ypoints)

font1 = {'family' : 'serif', 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'serif', 'color' : 'red', 'size' : 10}

plt.title("Sports Watch Data", loc='left', fontdict=font1)
plt.xlabel('No of Students', fontdict=font2)
plt.ylabel('Total Score',  fontdict=font2)

plt.show()
```

`Grid`

```python
xpoints = np.array([0,2,4,6,8]) 
ypoints = np.array([0,5,5,2,10]) 

plt.plot(xpoints, ypoints)

font1 = {'family' : 'serif', 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'serif', 'color' : 'red', 'size' : 10}

plt.title("Sports Watch Data", loc='left', fontdict=font1)
plt.xlabel('No of Students', fontdict=font2)
plt.ylabel('Total Score',  fontdict=font2)

plt.grid()
plt.show()
```

```python
xpoints = np.array([0,2,4,6,8]) 
ypoints = np.array([0,5,5,2,10]) 

plt.plot(xpoints, ypoints)

font1 = {'family' : 'serif', 'color' : 'blue', 'size' : 20}
font2 = {'family' : 'serif', 'color' : 'red', 'size' : 10}

plt.title("Sports Watch Data", loc='left', fontdict=font1)
plt.xlabel('No of Students', fontdict=font2)
plt.ylabel('Total Score',  fontdict=font2)

plt.grid(axis='x', linestyle=':', linewidth=2)
plt.show()
```

`Subplot`

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
#the figure has 2 row, 1 column, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
#the figure has 2 row, 1 column, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

### Histogram

A histogram is a graph showing `frequency distributions`.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

`Pie Charts`

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow=True)
plt.legend(title = "Four Fruits:")
plt.show() 
```
